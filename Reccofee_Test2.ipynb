```python
import pandas as pd
df = pd.read_csv('/content/coffee_fix.csv')
df.head()
```

```python
values = {'roast': 'Unknown', 'aroma': 0.0, 'acid': 0.0, 'body': 0.0, 'flavor': 0.0, 'aftertaste': 0.0}
df.fillna(values, inplace=True)
df
```

```python
data = df.filter(['name','aroma','acid','body'])
data
```

```python
X = data.loc[:, data.columns != 'name'].values
y = data['name']
y.head()
```

```python
# X = data.iloc[:, 0:-1]
# y = data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
```

```python
model.fit(X, y)
```

```python
model.score(X,y)
```

```python
predictions = model.predict(X)
predictions[:500]
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y, predictions))
print(confusion_matrix(y, predictions))
print(classification_report(y, predictions))
```

```python
import joblib

joblib.dump(model,'/content/randomforest_model.pkl')
```
