#**Import Library**

```python
# Data loading and data analysis
import numpy as np
import pandas as pd
import json
import zipfile
from google.colab import files

# Data visualization
import matplotlib.pyplot as plt

# Data preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

# Splitting data
from sklearn.model_selection import train_test_split

# Modelling
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
coffee = pd.read_csv('coffee.csv')
coffee.head(5)
```

```python
recom_data = coffee.filter(['name', 'aroma', 'acid', 'body', 'rating', 'roaster', 'desc_1', 'desc_2'])
values = {'roast': 'Unknown', 'aroma': 0.0, 'acid': 0.0, 'body': 0.0, 'flavor': 0.0, 'aftertaste': 0.0}
recom_data.fillna(values, inplace=True)
recom_data
```

```python
data = coffee.filter(['aroma', 'acid', 'body', 'name'])
data.head(5)
```

```python
data.rename(columns=({'name':'label'}), inplace=True)
data.head(5)
```

##**Data Analysis**

**Encode label**

```python
labelEncoder = LabelEncoder()

labelEncoder.fit_transform(data["label"])
labelEncoder_key_value = dict(zip(labelEncoder.classes_, labelEncoder.transform(labelEncoder.classes_)))

data_encode = pd.DataFrame(labelEncoder_key_value.items(), columns=['Original Label', 'Encoded'])
```

```python
data_encode.head(5)
```

```python
data['label'] = labelEncoder.transform(data['label'])
data
```

```python
X = data.iloc[:, 0:-1]
y = data.iloc[:, -1]
```

```python
model = svm.SVC()
model_2 = KNeighborsClassifier()
model_3 = RandomForestClassifier(n_estimators=100)

model.fit(X, y)
model_2.fit(X, y)
model_3.fit(X, y)
```
